## Step 1. Import the necessary libraries

```python
import pandas as pd
```

## Step 2. Import the dataset from this address.

## Step 3. Assign it to a variable called users

```python
users = pd.read_csv('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user', delimiter='|')
```

## Step 4. Discover what is the mean age per occupation


```python
mean_age= users[['occupation','age']].groupby('occupation').mean()
mean_age
```

## Step 5. Discover the Male ratio per occupation and sort it from the most to the least

```python
df=users[['gender','occupation']]
males = df.loc[df['gender']=='M']
group_occupations=df.groupby('occupation').count()
male_occupation_count=males.groupby('occupation').count()
male_ratio= male_occupation_count/group_occupations
male_ratio.sort_values(ascending=False,by='gender')
```

## Step 6. For each occupation, calculate the minimum and maximum ages


```python
df=users[['occupation','age']]
min_age= df.groupby('occupation').min()
max_age= df.groupby('occupation').max()
max_min_age=pd.concat([max_age,min_age],axis=1)
max_min_age
```

## Step 7. For each combination of occupation and sex, calculate the mean age

```python
df=users[['occupation','gender','age']]
mean_age_of_occupation_and_sex= df.groupby(['occupation','gender']).mean('age')
mean_age_of_occupation_and_sex
```

## Step 8. For each occupation present the percentage of women and men

```python
women_and_men_in_occupation= df.groupby('occupation')['gender'].value_counts(normalize=True)*100
women_and_men_in_occupation
```

# Wind statistics

## Step 1. Import the necessary libraries

```python
import pandas as pd
```

## Step 2. Import the dataset from the attached file wind.txt

```python
data=pd.read_csv('wind.txt',delimiter='\s+')
```

## Step 3. Assign it to a variable called data and replace the first 3 columns by a proper datetime index

```python
data.rename(columns={"Yr":"Year","Mo":"Month","Dy":"Day"},inplace=True)
data
```

## Step 4. Year 2061? Do we really have data from this year? Create a function to fix it and apply it.

```python
def fix_year():
    data['Year']+=1900
fix_year()

data
```

## Step 5. Set the right dates as the index. Pay attention at the data type, it should be datetime64[ns].

```python
data.index=pd.to_datetime(pd.concat([data['Year'],data['Month'],data['Day']],axis=1))
data.drop(['Year','Month','Day'],axis=1,inplace=True)
data
```

```python
data.index
```

## Step 6. Compute how many values are missing for each location over the entire record.They should be ignored in all calculations below.

```python
data.isna().sum()
```

```python
data.isnull().sum().sum()
```

## Step 7. Compute how many non-missing values there are in total.

```python
data.notnull().sum().sum()
```

## Step 8. Calculate the mean windspeeds of the windspeeds over all the locations and all the times.

```python
data.mean().mean()
```

## Step 9. Create a DataFrame called loc_stats and calculate the min, max and mean windspeeds and standard deviations of the windspeeds at each location over all the days. A different set of numbers for each location.

```python
loc_stats = pd.concat([data.min(), data.max(), data.mean(), data.std()], axis=1) 
loc_stats.rename(columns={0:'Min', 1:'Max', 2:'Mean', 3:'Std'}, inplace=True) 

loc_stats
```

## Step 10. Create a DataFrame called day_stats and calculate the min, max and mean windspeed and standard deviations of the windspeeds across all the locations at each day. A different set of numbers for each day.

```python
day_stats=pd.concat([data.min(axis=1),data.max(axis=1),data.mean(axis=1),data.std(axis=1)],axis=1)
day_stats.rename(columns={0:'Min',1:'Max',2:'Mean',3:'Std'},inplace=True)
day_stats
```

## Step 11. Find the average windspeed in January for each location.Treat January 1961 and January 1962 both as January.

```python
avg_windspeed=data[data.index.month==1].mean()
avg_windspeed
```

## Step 12. Downsample the record to a yearly frequency for each location

```python
yearly= data.resample('Y').ffill()
yearly
```

## Step 13. Downsample the record to a monthly frequency for each location.

```python
monthly= data.resample('M').ffill()
monthly
```

## Step 14. Downsample the record to a weekly frequency for each location.

```python
weekly= data.resample('W').ffill()
weekly
```

## Step 15. Calculate the min, max and mean windspeeds and standard deviations of the windspeeds across all locations for each week (assume that the first week starts on January 2 1961) for the first 52 weeks.

```python
weekly_stats = weekly[:52]
weekly_stats.shape[0]
```

```python
weekly_stats = pd.concat([weekly_stats.mean(), weekly_stats.max(), weekly_stats.mean(), weekly_stats.std()],axis=1)
weekly_stats.rename(columns={0:'Mean',1:'Max',2:'Mean',3:'Std'},inplace=True)
weekly_stats
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('us-marriages-divorces-1867-2014.csv')
df
```

```python
fig= plt.figure()
plt.rcParams["figure.figsize"] = [15, 10]
df.plot( 'Year', ['Marriages_per_1000','Divorces_per_1000'], label=['Marriages','Divorces'])
plt.title('Marriages and Divorces per capita in the U.S. between 1867 and 2014', color ='k', fontsize= 24)

plt.xlabel('Year', color ='k', fontsize = 16)
plt.ylabel('Capita', color ='k', fontsize = 16)

plt.show()
```

# Question 7

```python
filter_by_years = (df['Year'] == 1900) | (df['Year'] == 1950) | (df['Year'] == 2000) 
df2 = df.loc[filter_by_years, ['Year','Marriages_per_1000','Divorces_per_1000']]
df2
```

```python
df2[{'Marriages_per_1000','Divorces_per_1000'}].plot.bar()
plt.title("Marriages and Divorces per capita in the U.S. between 1900,1950 and 2000",color ='k',fontsize =24)
plt.xlabel('Year', color= 'k',fontsize = 18)
plt.ylabel('Capita', color= 'k',fontsize = 18)

plt.xticks((0, 1, 2), ("1900", "1950", "2000"), rotation = 0)
```
